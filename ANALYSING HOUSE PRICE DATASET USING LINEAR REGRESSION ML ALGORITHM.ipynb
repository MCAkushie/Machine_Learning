```python
# ANALYZING HOUSE PRICE DATASET USING LINEAR REGRESSION MACHINE LEARNING ALGORITHM

# importing relevant libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt


# sklearn package for machine learning in python:
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# setting up my data frame from the CSV file
df = pd.read_csv('houseprice_data.csv')

print(df.corr()['price'].sort_values(ascending=False))
```

```python
# setting X and y values from my data frame
X = df.iloc[:, [3]].values
y = df.iloc[:, 0].values
```

```python
# setting up train and set data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3,random_state = 0)
```

```python
# instantiating my model
regr = LinearRegression()
regr.fit(X_train, y_train)
```

```python
# evaluating our model with sklearn metrics
# The coefficients
print('Coefficients: ', regr.coef_)
# The intercept
print('Intercept: ', regr.intercept_)
# The mean squared error
print('Mean squared error: %.8f'% mean_squared_error(y_test, regr.predict(X_test)))
# The R^2 value:
print('Coefficient of determination: %.2f'% r2_score(y_test, regr.predict(X_test)))
```

```python
# Visualize the test data set
plt.subplot()
plt.scatter(X_test, y_test, color = 'black')
plt.plot(X_test, regr.predict(X_test), color = 'red')
plt.title('Test data: Price VS Sqft_living')
plt.xlabel('sqft_living')
plt.ylabel('Price')
```

```python
# Predicting some values
y_pred=regr.predict(X_test)
print(y_pred)
```

```python
# Plotting the Results
plt.figure()
plt.scatter(y_test,y_pred, color = 'black')
plt.xlabel('predicted Values')
plt.ylabel('Actual Values')
plt.title('Actual vs. Predicted')
```
